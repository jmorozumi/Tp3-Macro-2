```python
import numpy as np
from scipy.optimize import fsolve
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import fsolve

# Encontramos los valores de equilibrio, para luego poder graficar ambos graficos.
# Parámetros
beta = 0.5
gamma = 1 - beta
A = 1.5
K = 1
r = 0.04
b = 0.1
C = 1
L_bar = 100

# Definición de las funciones α(E) y a(E)
def alpha(E):
    return K**(1/gamma) * (b*E)**((gamma-1)/gamma) * (L_bar - E)**(beta/gamma)

def a(E):
    return b * E / (L_bar - E)

# Ecuación a resolver: rV_v(E) = -C + α(E)/(a(E) + α(E) + 2b + 2r) * A = 0
def equilibrium_equation(E):
    return -C + alpha(E) / (a(E) + alpha(E) + 2*b + 2*r) * A

# Encontrar el valor de E que satisface la ecuación
E_initial_guess = 50  # Suposición inicial para E
E_solution = fsolve(equilibrium_equation, E_initial_guess)[0]

# Calcular w, α(E) y a(E)
alpha_E = alpha(E_solution)
a_E = a(E_solution)
w = (a_E + b + r) * A / (a_E + alpha_E + r + b)

# Mostrar los resultados
print("El valor de E de equilibrio es:", E_solution)
print("El valor de α(E) es:", alpha_E)
print("El valor de a(E) es:", a_E)
print("El valor del salario (w) es:", w)
```

```python
#Gráfico de equilibrio de salario en el modelo DMP
# Asumir valores para las constantes
a = 0.58
b = 0.1
r = 0.04
A = 1.5
alpha = 1.72

# Crear un rango de valores para w
w_values = np.linspace(0, 1, 100)

# Calcular las funciones
Ve_minus_Vu = w_values / (a + b + r)
Vf_minus_Vv = (A - w_values) / (alpha + b + r)

# Graficar las funciones
plt.figure(figsize=(10, 6))
plt.plot(w_values, Ve_minus_Vu, label=r'$Ve - Vu = \frac{w}{a + b + r}$')
plt.plot(w_values, Vf_minus_Vv, label=r'$Vf - Vv = \frac{A - w}{alpha + b + r}$')
plt.xlabel('SALARIOS')
plt.ylabel('Beneficios')
plt.title('Salario de Equilibrio Modelo DMP')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Gráfico del empleo en equilibrio en el modelo DMP
#Parámetros y funciones supuestas
def alpha(E):
    return (100 - E) / (0.1 * E) 

def a(E):
    return 0.1 * E /(100 - E)  

b = 0.1
r = 0.04
C = 1
E = 85

# Valores de A
A = 1.5

# Valores de empleo
E = np.linspace(0, 100, 500)

# Función rV_V(E)
def rVV(E, A):
    return -C + (alpha(E) / (a(E) + alpha(E) + 2*b + 2*r)) * A

# Cálculo de rV_V(E) para A.
rVV = rVV(E, A)

# Crear la gráfica
plt.figure(figsize=(10, 6))
plt.plot(E, rVV, label='A', color='blue')
plt.axhline(0, color='gray', linewidth=0.5)
plt.title('Equilibrio Modelo DMP')
plt.xlabel('Empleo')
plt.ylabel('rV_V')
plt.legend()
plt.grid(True)
plt.show()
```
