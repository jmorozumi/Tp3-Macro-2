```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import fsolve
```

```python
#Suponemos el shock de la subida de A de 1,5 a 2 
#Parámetros y funciones supuestas
def alpha(E):
    return (100 - E) / (0.1 * E) 

def a(E):
    return 0.1 * E /(100 - E)  

b = 0.1
r = 0.04
C = 1
E = 85

# Valores de A para alto y bajo
A_alto = 1.5
A_bajo = 2

# Valores de empleo
E = np.linspace(0, 100, 500)

# Función rV_V(E)
def rVV(E, A):
    return -C + (alpha(E) / (a(E) + alpha(E) + 2*b + 2*r)) * A

# Cálculo de rV_V(E) para A alto y A bajo
rVV_alto = rVV(E, A_alto)
rVV_bajo = rVV(E, A_bajo)

# Crear la gráfica
plt.figure(figsize=(10, 6))
plt.plot(E, rVV_alto, label='A bajo', color='blue')
plt.plot(E, rVV_bajo, label='A alto', color='red')
plt.axhline(0, color='gray', linewidth=0.5)
plt.title('Equilibrio Modelo DMP')
plt.xlabel('Empleo')
plt.ylabel('rV_V')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Parámetros constantes
a = 0.2
b = 0.1
r = 0.04
A_inicial = 1.5  # Valor de A antes del shock
A_shock = 2.0    # Valor de A después del shock
w = np.linspace(0, 2, 500)  # Rango de valores del salario

# Definir las funciones
def Ve_Vu(w, A):
    return w / (a + b + r)

def Vf_Vv(w, A):
    return (A - w) / (a + b + r)

# Calcular los valores de las funciones antes y después del shock
Ve_Vu_inicial = Ve_Vu(w, A_inicial)
Vf_Vv_inicial = Vf_Vv(w, A_inicial)
Ve_Vu_shock = Ve_Vu(w, A_shock)
Vf_Vv_shock = Vf_Vv(w, A_shock)

# Graficar los resultados
plt.figure(figsize=(10, 6))

plt.plot(w, Ve_Vu_inicial, label=r'$Ve - Vu$ (A inicial)', color='blue')
plt.plot(w, Vf_Vv_inicial, label=r'$Vf - Vv$ (A inicial)', color='red')
plt.plot(w, Vf_Vv_shock, label=r'$Vf - Vv$ (A alto)', color='red', linestyle='dotted')

plt.xlabel('Salario')
plt.ylabel('Beneficios')
plt.title('Salario de Equilibrio Modelo DMP')
plt.legend()
plt.grid(True)
plt.show()
```
